### Exercício 1: Registro de Pacientes

Crie uma classe chamada Paciente que represente os pacientes de uma clínica de saúde. A classe deve ter as seguintes propriedades:

nome: Nome do paciente
idade: Idade do paciente
sexo: Sexo do paciente (masculino/feminino)
historico_medico: Lista que armazena o histórico médico do paciente

Além disso, a classe deve ter os seguintes métodos:

adicionar_historico(medico, diagnostico): Método que adiciona uma entrada ao histórico médico do paciente, contendo o nome do médico e o diagnóstico fornecido.
imprimir_historico(): Método que imprime o histórico médico do paciente.

```python

class Paciente:

    def __init__(self, nome, idade, genero) :
        self.nome = nome
        self.idade = idade
        self.genero = genero
        self.historico = []


    def Add_Historico(self, medico, diagnostico):
        self.medico = medico
        self.diagnostico = diagnostico
        self.historico.append({
            'Medico': self.medico,
            'Diagnostico': self.diagnostico
            })
    
    def Imp_Historico(self):
        print(f'{10*"-"}Historico{9*"-"}')
        for i in range(len(self.historico)):
            print(f"""
Nome: {self.nome}
Idade: {self.idade}
Genero: {self.genero}
Medico: {self.historico[i]['Medico']}
Diagnostico: {self.historico[i]['Diagnostico']}
{30*'_'}
            """)


paciente_1 = Paciente(nome='Gabriel', idade='22', genero='masculino')

diagnostico_1 ='primeiro diagnostico'

paciente_1.Add_Historico(medico='Dr.Hans Shucrute', diagnostico=diagnostico_1)

diagnostico_2 = 'segundo diagnostico'

paciente_1.Add_Historico(medico='Dr.Hans Shucrute', diagnostico=diagnostico_2)

paciente_1.Imp_Historico()
```

### Exercício 2: Controle de Equipamentos Médicos

Crie uma classe chamada EquipamentoMedico que representa os equipamentos médicos de uma clínica. A classe deve ter as seguintes propriedades:

nome: Nome do equipamento
tipo: Tipo de equipamento (ex: "Raio-X", "Tomógrafo", etc.)
status: Status do equipamento (disponível/manutenção/fora de serviço)
Além disso, a classe deve ter os seguintes métodos:

alterar_status(novo_status): Método que altera o status do equipamento.
verificar_status(): Método que retorna o status atual do equipamento.

```python
class Equip_Medico:
    def __init__(self, nome, tipo, status):
        self.nome = nome
        self.tipo = tipo
        self.status = status

    def Alt_Status(self, status):
        self.status = status
        print(f'O status do equipamento foi alterado com exito!')
    
    def Verf_Status(self):
        print(f'O status do equipamento {self.nome} é: {self.status}')

equip_1 = Equip_Medico(nome='RX-1', tipo='Equipamento de raio-x', status='disponivel')
equip_2 = Equip_Medico(nome='RX-2', tipo='Equipamento de raio-x', status='fora de serviço')
equip_3 = Equip_Medico(nome='RE-1', tipo='Respirador', status='disponivel')
equip_4 = Equip_Medico(nome='DF-1', tipo='Desfibrilador', status='manutenção')

equip_1.Verf_Status()

equip_4.Verf_Status()
equip_4.Alt_Status('disponivel')
equip_4.Verf_Status()
```

### Exercício: Registro de Consultas Médicas
Crie uma classe chamada ConsultaMedica que represente as consultas médicas realizadas em uma clínica. A classe deve ter as seguintes propriedades:

data: Data da consulta

paciente: Objeto da classe Paciente representando o paciente da consulta

medico: Nome do médico responsável pela consulta

diagnostico: Diagnóstico fornecido durante a consulta

Além disso, a classe deve ter um método:

imprimir_informacoes(): Método que imprime as informações da consulta.

```python
import datetime as dt

class ConsultaMedica:
    def __init__(self, paciente, diagnostico):
        self.paciente = paciente
        self.data = dt.datetime.today()
        self.diagnostico = diagnostico

    def Imp_Info(self):
        print(f"""
{11*'-'}Consulta{11*'-'}
Paciente: {self.paciente}
Data: {self.data}
Diagnostico: {self.diagnostico}
        """)

consulta_1 = ConsultaMedica(paciente='Gabriel', diagnostico='Tuberculose')
consulta_2 = ConsultaMedica(paciente='Fenando', diagnostico='Morreu de morte morrida, mas passa bem.')

consulta_1.Imp_Info()
consulta_2.Imp_Info()
```
