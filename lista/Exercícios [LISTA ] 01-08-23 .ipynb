### 1 - Crie uma lista contendo os nomes de 5 países. Peça ao usuário para inserir o nome de mais um país. Em seguida, verifique se o país fornecido pelo usuário está na lista e exiba uma mensagem informando se ele está ou não na lista.

```python
paises = ['Gabão', 'Zamia', 'Georgia', 'Monaco', 'Pais Basco']

select = input('Insira o nome de um pais').title()

if select in paises :
    print('{} está na lista'.format(select))
else :
    print('{} não está na lista'.format(select))
```

### 2 - Crie uma lista contendo os preços de 5 produtos. Em seguida, calcule e exiba o preço médio desses produtos.

```python
preços = [400, 250, 40, 79, 11000]

media = sum(preços) / len(preços)

print(media)
```

```python
cores = ['vermelho', 'amarelo', 'azul', 'preto', 'branco']

select = input('selecione uma cor')

if select in cores :
    cores.remove(select)
    print('A cor {} foi removida, segue abaixo as cores diponiveis \n {}'.format(select, cores))
else :
    print('A cor {} não esta na lista!'.format(select))
```
