```python
import time
def validaNumero(numero, listaCandidatos) :
    
    if numero in listaCandidatos :
        return 0
    elif numero == '' :
        return 1
    elif numero == 'SAIR123!' or numero == 'sair123!':
        return 2
    elif numero.isnumeric() != True :
        return 3
    else :
        return 4

def validaNome(nome):
    
    for i in '/_*@:;?!´`' :
        if i in nome :
            return 0
        elif i.isnumeric() == True :
            return 1
    if nome == 'senha123' :
        return 2
    elif nome == '' :
        return 3
    else :
        return 4    
    
def validaPartido(partido) :
    if partido == 'senha123' :
        return 0
    elif partido == '' :
        return 1
    else :
        return 2


    
def cadastrarCandidatos():
    print(f'{20 * "-"}Cadastro de Candidatos{20 * "-"} \n *Se quiser sair da pagina de cadastro, digite "SAIR123!".')
    while True:
        # try :
            numero = input('Digite o numero do candidato: ')
            match validaNumero(numero, candidatos) :
                case 0 :
                    print(f'Numero "{numero}" já foi registrado!!!')
                    continue
                case 1 :
                    print('O campo precisa ser preenchido!')
                    continue
                case 2 :
                    print('!!!!!!!CADASTRAMEMTO ENCERRADO!!!!!!!')
                    for candidato in candidatos :
                        print(f'{candidato} -- Nome: {candidatos[candidato]["nome"]}, Partido: {candidatos[candidato]["partido"]}')

                case 3 :
                    print('Digite apenas numeros!')
                case 4 :
                    while True:
                        nome = input('Digite o nome do candidato: ').strip()
                        
                        match validaNome(nome) :
                            case 0 :
                                print('Não é permitido caracteres especiais alem de "-" e "."')
                                continue
                            case 1 :
                                print('Digite um Nome valido!')
                                continue
                            case 2 :
                                break
                            case 3 :
                                print('O campo precisa ser preenchido!')
                            case 4 :
                                while True :
                                    partido = input('Digite o nome do Partido.')

                                    match validaPartido(partido) :
                                        case 0 :
                                            break
                                        case 1 :
                                            print('O campo precisa ser preenchido')
                                            continue
                                        case 2 :
                                            candidatos.update({numero : {'nome' : nome.upper(),'partido' : partido.upper(), 'votos' : 0}})
                                            print('Cadatro realizado com sucesso!!!')
                                    break
                                break
                    continue

        # except :
        #     print('\n !!!!Erro!!!! \n Não foi possivel realizar o cadastro, tente novamente!!\n')
            



def validaVoto(voto,listaCandidatos) :
    if voto == 'senha123':
        return 0
    elif voto == '' :
        return 1
    
    elif listaCandidatos.get(voto) == None :
        return 2
    else :
        return 3

def confirmaVoto(escolha) :
   
    while True :
        confirmar = input('Confirma? - ("s" ou "n")')
        if confirmar == 's' :
            candidatos['total'] += 1
            candidatos [escolha] += 1
            print('Voto Confirmado!!')
            time.sleep(3)
            break
        elif confirmar == 'n' :
            break
        else :
            print('Responda com "s" ou "n"!')
            continue

def mostraCandidato(voto) :
    nome = candidatos[voto] ['nome']
    partido = candidatos[voto] ['partido']

    return print(f'\n Numero: {voto}\nCandidato: {nome}\nPartido: {partido}')

def iniciaVotação() :
    if len(candidatos) < 2 :
        return print('Precisa ter mais que 2 candidatos para começar a votação!!')
    else :
        while True :
            try:
                voto = input('Digite o numero do candidato: ')

                match validaVoto(voto, candidatos) :
                    case 0 :
                        break
                    case 1 :
                        print('O campo precisa ser preenchido!')
                        continue
                    case 2 :
                        print('Candidato não encontado!')
                        confirmaVoto('nulo')
                        break
                    case 3 :
                        mostraCandidato(voto)
                        confirmaVoto(candidatos[voto]['votos'])
                        break
                        
            except:
                print('\n !!Erro!! \n Chame o mesario')
                

def mostraResultado() :
    if candidatos['total'] < 3 :
        return print('Não é possivel mostrar o resultado da eleição pois foram computados menos de 3 votos!!')
    c = 0
    for itens in candidatos[3:] :
        if itens['votos'] == 0 :
            c += 1
    if len(candidatos) < c :
        return print('Os candidatos precisam ter ao menos 1 voto!')
    else :
        
        print(f'{20*"-"}Resultados das eleições{20*"-"}')
        i = 10
        time.sleep(10)
        for candidato in candidatos[3:] :
            print(f"""
{mostraCandidato(candidato)}
Votos: {candidatos[candidato]['votos']}
{100 * '-'}
""")
        

    
    
candidatos = {'total': 0, 'nulo':0, 'branco': 0}

def menu() :
    print(f"{20 * '-'}'URNA ELETRONICA'{20 * '-'}\n\n 1- Cadastrar candidatos\n 2-Iniciar eleição\n 3-Resultados da eleição!\n 4- Sair\n\n\n ")
    # try :
    while True :
        opçao = input('Escolha uma opção: ')
        match opçao :
            case '1' :
                cadastrarCandidatos()
            case '2' :
                iniciaVotação()
            case '3' :
                mostraResultado()
            case '4' :
                break
    # except :
    #     print('deu ruim')
        

menu() 
```
