## Tarefas:

## 1 - Registro de Empréstimos:

## 2 -  Avaliação de Livros:

## 3 - Avaliação do Atendimento da Biblioteca:

## 4 - Imprimir Relatório:

### Após realizar as ações anteriores, imprima um relatório que apresente as informações sobre os empréstimos, avaliações dos livros e a média das avaliações do atendimento da biblioteca.


```python
# codigo aqui

biblioteca = {'lord Of the rings' : {'avaliação' : 'não há avaliações', 'status' : 'disponivel', 'Nº de emprestimos': 0},
               'game Of thrones' : {'avaliação' : 'não há avaliações', 'status' : 'disponivel', 'Nº de emprestimos': 0},
               'harry potter' : {'avaliação' : 'não há avaliações', 'status' : 'disponivel', 'Nº de emprestimos': 0},
               'hunger games' : {'avaliação' : 'não há avaliações', 'status' : 'disponivel', 'Nº de emprestimos': 0}
            }

while True :
    menu = input('Secelcione uma das seguintes opções:\n 1- Exibir Livros\n 2- Emprestar livro\n 3- Retronar livro\n 0- Sair do menu')
    match menu :
        case '0' :
            break
        case '1' :
            for i, livro in enumerate(biblioteca) :
                print(f'{i+1}-{livro}: \n---Avaliação: {biblioteca[livro]["avaliação"]}\n--Avaliação do livro: {biblioteca[livro]["avaliação"]}\n-- Nº de emprestimos: {biblioteca[livro]["Nº de emprestimos"]}')

            print('*' * 60)

        case _ :
            break
```
