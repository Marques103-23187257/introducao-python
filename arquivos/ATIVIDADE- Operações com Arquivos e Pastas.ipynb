#### Exercício 4:
Crie um programa que verifique se um arquivo existe em um diretório usando os.path.exists()

```python
from os import path

verifica = lambda pasta : print('A pasta existe') if path.exists(pasta) == True else print('A pasta não existe')

verifica('nova pasta')
```

#### Exercício 6:
Crie um programa que escreva uma lista de strings em um arquivo de texto usando open()

```python
lista = ['iphone', 'galaxy', 'motorola', 'bastemp', 'tramontina']
arq = open('lista.txt', 'w')
for item in lista :
    arq.write(item + '\n')
arq.close()
arq = open('lista.txt', 'r')
conteudo = arq.read()
print(conteudo)
arq.close()
```

#### Exercício 7:
Crie um programa que copie um arquivo de um diretório para outro usando shutil.copy()

```python
from shutil import copy
from shutil import SameFileError

arq = 'lista.txt'
destino = 'nova pasta/lista.txt' 

try :
    copy(arq, destino)
    print('Arquivo copiado com sucesso')
except SameFileError :
    print('O arquivo ja existe')
```

#### Exercício 8:
Crie um programa que renomeie um arquivo usando os.rename()

```python
from os import rename

arq = 'nova pasta/lista.txt'

novo_nome = 'nova pasta/lista123.txt'
try :
    rename(arq, novo_nome)
    print('O arquivo foi renomeado com suceso!')
except :
    print('naõ foi possivel renomear o arquivo')
```

#### Exercício 9:

#### Exercício 10:
Crie um programa que exclua um diretório e todo o seu conteúdo usando a biblioteca pathlib com o método rmdir()

```python
from shutil import rmtree

caminho = 'nova pasta'
try:
    rmtree(caminho)
    print('pasta exluida com sucesso!')
except :
    print('não foi possivel excluir a pasta!')
```

#### Exercício 11:
Crie um programa que liste todos os arquivos em um diretório específico usando a biblioteca pathlib com o método iterdir()

```python
from pathlib import Path

diretorio = Path('c:/Users/Aluno 11/desktop/python')

for arquivo in diretorio.iterdir():
    if arquivo.is_file():
        print(arquivo.name)
```
